### 📊 ANÁLISE E PRÉ-PROCESSAMENTO DE DADOS — Dataset Titanic

#### 🔹 Objetivo

Preparar o dataset para análise e modelagem, garantindo que os dados estejam limpos, consistentes e utilizáveis.

#### 🔹 Etapas do pré-processamento

* “Arrumar” o dataset
* Remover dados não utilizáveis
* Obter primeiras impressões
* Tratar dados incorretos ou ausentes

#### 📁 Dataset utilizado

* Titanic

```python
import pandas as pd
import numpy as np

# Dataset público do Titanic
url = "https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv"

df = pd.read_csv(url)

df.head()
```

#### 🧹 Remoção de colunas irrelevantes

Algumas colunas não agregam valor para análise, como identificadores artificiais.

##### ❌ Exemplo: coluna `PassengerId`

```python
# Removendo a coluna PassengerId
df = df.drop('PassengerId', axis=1)

# Visualizando informações do dataset
df.info()

# A coluna `PassengerId` é apenas um identificador único e não contribui para análise ou previsão.
```

#### ⚠️ Tratamento de dados nulos

Dados ausentes podem prejudicar análises e modelos. Existem três abordagens principais:

* Substituir por valores (média, mediana, etc.)
* Remover colunas
* Remover linhas

#### 👶 Preenchendo valores nulos na coluna `Age`

```python
# Preenchendo idades nulas com a mediana por classe e sexo
df['Age'] = df.groupby(['Pclass', 'Sex'], group_keys=False)['Age'] \
              .apply(lambda x: x.fillna(x.median()))

# Agrupa os dados por `Pclass` (classe) e `Sex` (sexo)
# Calcula a **mediana da idade** para cada grupo
# Substitui os valores nulos com essa mediana
```

#### 🚫 Removendo coluna com muitos valores nulos (`Cabin`)

```python
# Verificando quantidade de valores nulos
df['Cabin'].isnull().sum()
# Resultado esperado: 687

# Removendo a coluna Cabin
df = df.drop('Cabin', axis=1)

# A coluna `Cabin` possui muitos valores ausentes, o que dificulta sua utilização.
```

#### 🧾 Removendo linhas com valores nulos (`Embarked`)

```python
# Verificando valores nulos
df['Embarked'].isnull().sum()
# Resultado esperado: 2

# Removendo linhas com valores nulos
df = df.dropna(subset=['Embarked'])

# `df['Embarked'].isnull().sum()` → Conta quantos valores estão faltando na coluna
# `df.dropna(subset=['Embarked'])` → Remove apenas as linhas onde `Embarked` está nulo
```

#### 🔥 Heatmap de Correlação

* Analisar a correlação entre variáveis numéricas
* Identificar relações fortes entre variáveis
* Detectar possíveis redundâncias (dados muito correlacionados)

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Selecionando apenas variáveis numéricas
df_num = df.select_dtypes(include=['int64', 'float64'])

# Matriz de correlação
corr = df_num.corr()

# Plotando heatmap
plt.figure(figsize=(10,6))
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Heatmap de Correlação')
plt.show()

# Valores próximos de **1** → forte correlação positiva
# Valores próximos de **-1** → forte correlação negativa
# Valores próximos de **0** → pouca ou nenhuma relação

# Se duas variáveis forem muito correlacionadas, pode indicar: redundância e possível remoção em modelos futuros
```

#### 📦 Boxplot (Distribuição e Outliers)

* Visualizar a distribuição dos dados
* Identificar quartis
* Detectar outliers (valores muito distantes)

##### 📌 Boxplot da variável `Age`

```python
plt.figure(figsize=(6,4))
sns.boxplot(x=df['Age'])
plt.title('Boxplot - Age')
plt.show()
```

##### 📌 Boxplot da variável `Fare`

```python
plt.figure(figsize=(6,4))
sns.boxplot(x=df['Fare'])
plt.title('Boxplot - Fare')
plt.show()
```
