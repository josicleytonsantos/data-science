### Funções

#### 📌 O que é uma função?

Uma função é um bloco de código reutilizável que recebe **entradas (parâmetros)** e retorna uma **saída (resultado)**.

#### 📌 Por que usar funções?

* Reutilizar código
* Melhorar organização
* Facilitar manutenção

```python
    def nome_funcao(parametros):
        # corpo da função
        return resultado
```

* `def`: define a função
* `parametros`: opcionais
* `return`: opcional (valor retornado)

````python
def soma(a, b):
    return a + b

resultado = soma(2, 3)
print(resultado)  # 5
````

````python
#Parâmetros com valor padrão

def saudacao(nome="Usuário"):
    return f"Olá, {nome}!"

print(saudacao())          # Olá, Usuário!
print(saudacao("João"))    # Olá, João!
````

### Objetos Mutáveis vs Imutáveis

#### 📌 Mutáveis

Podem ser alterados após criação:

* lista (`list`)
* dicionário (`dict`)

#### 📌 Imutáveis

Não podem ser alterados:

* inteiro (`int`)
* float (`float`)
* tupla (`tuple`)
* string (`str`)

````python
def adicionar_numero(lista, numero):
    # Criamos uma cópia para NÃO alterar a lista original
    nova_lista = lista.copy()
    
    y = 2 * numero
    nova_lista.append(y)
    
    return nova_lista

lista = [1, 2, 3]

nova_lista = adicionar_numero(lista, 5)

print(lista)        # [1, 2, 3]
print(nova_lista)   # [1, 2, 3, 10]
````

#### 💡 Por que usar `.copy()`?

Sem isso, você modificaria a lista original, pois listas são **mutáveis** e passadas por referência.

### Programação Orientada a Objetos (POO)

#### 📌 O que é?

Um paradigma baseado em:

* **Classes**
* **Objetos**

#### 📌 Conceitos principais

* **Classe**: modelo/estrutura
* **Objeto**: instância da classe
* **Atributos**: características (variáveis)
* **Métodos**: ações (funções da classe)

#### 🧱 Estrutura de uma classe

```python
    class NomeClasse:
        def __init__(self, atributos):
            # construtor
    
        def metodo(self):
            pass
```

````python
class Carro:
    def __init__(self, cor, nome, modelo):
        self.cor = cor
        self.nome = nome
        self.modelo = modelo

    def acelerar(self):
        print("Acelerando...")

    def frear(self):
        print("Freando...")


carro1 = Carro("Azul", "Fusca", "Volkswagen")

carro1.acelerar()
carro1.frear()

print(carro1.nome)  # Fusca
````

### Herança

#### 📌 O que é?

Permite que uma classe (filha) herde características de outra (pai).

#### 🧠 Conceitos

* **Classe pai (superclasse)**
* **Classe filha (subclasse)**
* Reutilização de código
* Possibilidade de sobrescrever métodos

````python
class Animal:
    def __init__(self, nome):
        self.nome = nome

    def falar(self):
        raise NotImplementedError("Este método deve ser implementado pela classe filha")


class Cachorro(Animal):
    def falar(self):
        print(f"{self.nome} está latindo!")

    def dormir(self):
        print(f"{self.nome} está dormindo!")


class Gato(Animal):
    def falar(self):
        print(f"{self.nome} está miando!")


# Testando

cachorro = Cachorro("Scooby")
print(cachorro.nome)
cachorro.falar()
cachorro.dormir()

print("------")

gato = Gato("Garfield")
print(gato.nome)
gato.falar()

# gato.dormir()  # ERRO: método não existe na classe Gato
````

Exercício 1: Faça uma função que dado um vetor de inteiros retorne o maior inteiro do vetor.

````python
# Vetor de inteiro é uma lista que armazena números inteiros. Podemos criar uma função para encontrar o maior valor em um vetor de inteiros.

def maiorValor(vetor):
    if not vetor:
        return None  # Retorna None se o vetor estiver vazio

    maior = vetor[0]  # Assume o primeiro elemento como o maior inicialmente

    for numero in vetor:
        if numero > maior:
            maior = numero  # Atualiza o maior se encontrar um número maior

    return maior

numeros = [3, 7, 2, 9, 5]
resultado = maiorValor(numeros)
print(f"O maior valor é: {resultado}")  # O maior valor é: 9
````

Exercício 2: Você está organizando uma lista de alunos por turmas em uma escola. Cada turma contém vários alunos, e cada aluno deve conter um nome, idade e id.

(Exercício da aula passada, porém deve ser resolvido utilizando o que aprendemos sobre POO.)

a) Criar as turmas: Crie duas turmas, cada uma contendo 2 alunos.

b) Adicionar um novo aluno: Adicione um novo aluno em uma das turmas.

c) Remover um aluno: Remova um aluno de uma das turmas.

d) Exibir os alunos de uma turma.

````python
class Aluno:
    def __init__(self, nome, idade, id):
        self.nome = nome
        self.idade = idade
        self.id = id

class Turma():
    def __init__(self, nome):
        self.nome = nome
        self.alunos = []

    def adicionar_aluno(self, aluno):
        self.alunos.append(aluno)

    def remover_aluno(self, id):
        for aluno in self.alunos:
            if aluno.id == id:
                self.alunos.remove(aluno)
                return True

    def listar_alunos(self):
        if not self.alunos:
            print("Nenhum aluno na turma.")
            print("Esta turma não possui alunos cadastrados.")
            
            return None
        
        for aluno in self.alunos:
            print(f"Nome: {aluno.nome}, Idade: {aluno.idade}, ID: {aluno.id}")


# (a) e (b) criação de turmas e alunos, e adição de alunos às turmas
turma1 = Turma("Turma A")
turma1.adicionar_aluno(Aluno("Alice", 20, 1))
turma1.adicionar_aluno(Aluno("Bob", 22, 2))

turma2 = Turma("Turma B")
turma2.adicionar_aluno(Aluno("Charlie", 21, 3))
turma2.adicionar_aluno(Aluno("David", 23, 4))

# (c) e (d) remover um aluno de uma turma e listar os alunos restantes
turma1.remover_aluno(1)  # Remove Alice da Turma A
print("Alunos na Turma A:")
turma1.listar_alunos()  # Deve listar apenas Bob
````

Exercício 3: Crie uma função que dado dois vetores retorne um novo vetor que é a concatenação de ambos os vetores

````python
def concatenar_vetores(vetor1, vetor2):
    
    novoVetor = vetor1.copy()  # Criamos uma cópia do primeiro vetor para não alterar o original
    for elemento in vetor2:
        novoVetor.append(elemento)  # Adicionamos cada elemento do segundo vetor ao novo vetor
    
    return novoVetor

vetor_a = [1, 2, 3]
vetor_b = [4, 5, 6]

resultado_concatenacao = concatenar_vetores(vetor_a, vetor_b)
print(resultado_concatenacao)  # [1, 2, 3, 4, 5, 6]
````

Exercício 4: Crie uma função que dados dois vetores retorne um novo vcetor que é composto dos maiores elementos de cada vetor

````python
def maiores_elementos(vetor1, vetor2):
    if not vetor1 or not vetor2:
        return None  # Retorna None se algum dos vetores estiver vazio

    maior_vetor1 = max(vetor1)
    maior_vetor2 = max(vetor2)

    return [maior_vetor1, maior_vetor2]

vetor_c = [7, 8, 9]
vetor_d = [10, 11, 12]

resultado_maiores = maiores_elementos(vetor_c, vetor_d)
print(resultado_maiores)  # [9, 12]
````

Exercício 5: Crie uma função que receba uma lista de palavras e retorne o número de palavras que tenham um tamanho maior que 5

````python
def contar_palavras_longas(lista_palavras):
    contador = 0
    for palavra in lista_palavras:
        if len(palavra) > 5:
            contador += 1
    return contador

palavras = ["casa", "computador", "gato", "programação", "python"]
resultado_contagem = contar_palavras_longas(palavras)
print(resultado_contagem)  # 3
````
