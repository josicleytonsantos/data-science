### Funções

#### 📌 O que é uma função?

Uma função é um bloco de código reutilizável que recebe **entradas (parâmetros)** e retorna uma **saída (resultado)**.

#### 📌 Por que usar funções?

* Reutilizar código
* Melhorar organização
* Facilitar manutenção

```python
    def nome_funcao(parametros):
        # corpo da função
        return resultado
```

* `def`: define a função
* `parametros`: opcionais
* `return`: opcional (valor retornado)

````python
def soma(a, b):
    return a + b

resultado = soma(2, 3)
print(resultado)  # 5
````

````python
#Parâmetros com valor padrão

def saudacao(nome="Usuário"):
    return f"Olá, {nome}!"

print(saudacao())          # Olá, Usuário!
print(saudacao("João"))    # Olá, João!
````

### Objetos Mutáveis vs Imutáveis

#### 📌 Mutáveis

Podem ser alterados após criação:

* lista (`list`)
* dicionário (`dict`)

#### 📌 Imutáveis

Não podem ser alterados:

* inteiro (`int`)
* float (`float`)
* tupla (`tuple`)
* string (`str`)

````python
def adicionar_numero(lista, numero):
    # Criamos uma cópia para NÃO alterar a lista original
    nova_lista = lista.copy()
    
    y = 2 * numero
    nova_lista.append(y)
    
    return nova_lista

lista = [1, 2, 3]

nova_lista = adicionar_numero(lista, 5)

print(lista)        # [1, 2, 3]
print(nova_lista)   # [1, 2, 3, 10]
````

#### 💡 Por que usar `.copy()`?

Sem isso, você modificaria a lista original, pois listas são **mutáveis** e passadas por referência.

### Programação Orientada a Objetos (POO)

#### 📌 O que é?

Um paradigma baseado em:

* **Classes**
* **Objetos**

#### 📌 Conceitos principais

* **Classe**: modelo/estrutura
* **Objeto**: instância da classe
* **Atributos**: características (variáveis)
* **Métodos**: ações (funções da classe)

#### 🧱 Estrutura de uma classe

```python
    class NomeClasse:
        def __init__(self, atributos):
            # construtor
    
        def metodo(self):
            pass
```

````python
class Carro:
    def __init__(self, cor, nome, modelo):
        self.cor = cor
        self.nome = nome
        self.modelo = modelo

    def acelerar(self):
        print("Acelerando...")

    def frear(self):
        print("Freando...")


carro1 = Carro("Azul", "Fusca", "Volkswagen")

carro1.acelerar()
carro1.frear()

print(carro1.nome)  # Fusca
````

### Herança

#### 📌 O que é?

Permite que uma classe (filha) herde características de outra (pai).

#### 🧠 Conceitos

* **Classe pai (superclasse)**
* **Classe filha (subclasse)**
* Reutilização de código
* Possibilidade de sobrescrever métodos

````python
class Animal:
    def __init__(self, nome):
        self.nome = nome

    def falar(self):
        raise NotImplementedError("Este método deve ser implementado pela classe filha")


class Cachorro(Animal):
    def falar(self):
        print(f"{self.nome} está latindo!")

    def dormir(self):
        print(f"{self.nome} está dormindo!")


class Gato(Animal):
    def falar(self):
        print(f"{self.nome} está miando!")


# Testando

cachorro = Cachorro("Scooby")
print(cachorro.nome)
cachorro.falar()
cachorro.dormir()

print("------")

gato = Gato("Garfield")
print(gato.nome)
gato.falar()

# gato.dormir()  # ERRO: método não existe na classe Gato
````
