### 📦 Variáveis em Python

#### 📌 O que são variáveis?

Variáveis são utilizadas para **armazenar informações na memória**.

- Cada variável armazena um valor
- Esse valor possui um **tipo de dado**

#### 🔢 Tipos de dados básicos

Alguns dos principais tipos em Python:

- `int` → números inteiros  
- `float` → números decimais  
- `str` → textos (strings)  
- `bool` → valores lógicos (`True` ou `False`)  
- `None` → ausência de valor  

#### 🤖 Tipagem em Python

Python é uma linguagem de **tipagem dinâmica**, ou seja:

- Você **não precisa declarar o tipo** da variável
- O próprio Python **identifica automaticamente** o tipo com base no valor

````python
idade = 25        # int
altura = 1.75     # float
nome = "Maria"    # str
ativo = True      # bool
````

````python
idade: int = "vinte e cinco"  # Python permite, mas está errado conceitualmente
````

### 🔀 Estruturas de Decisão

As **estruturas de decisão** permitem que o código **execute ações diferentes dependendo de uma condição**.

👉 Ou seja: um bloco de código **só será executado se uma condição for verdadeira**.

⚠️ Importante
- As condições devem resultar em True ou False
- A indentação define o que pertence a cada bloco
- Apenas o primeiro bloco verdadeiro será executado

````python
x = 10

if (x > 10):
    print("X maior que 10")
else:
    print("X menor ou igual a 10")
print("Essa mensagem não depende de x") # importância da indentação
````

### 🔀 Usando elif (senão se)

````python
nota = 7

if nota >= 9:
    print("Excelente")
elif nota >= 6:
    print("Aprovado")
else:
    print("Reprovado")
````

### ⚙️ Operadores Importantes em Estruturas de Decisão

#### 📖 Explicação

Os operadores são símbolos ou palavras que permitem comparar valores e combinar condições dentro das estruturas de decisão.

#### 🔎 Operadores de comparação

Eles comparam dois valores e retornam `True` ou `False`:

| Operador | Descrição                | Exemplo           | Resultado                        |
|----------|-------------------------|-------------------|---------------------------------|
| `==`     | Igual a                 | `x == 10`         | `True` se x for 10              |
| `!=`     | Diferente de            | `x != 5`          | `True` se x não for 5           |
| `>`      | Maior que               | `x > 3`           | `True` se x for maior que 3     |
| `<`      | Menor que               | `x < 8`           | `True` se x for menor que 8     |
| `>=`     | Maior ou igual a        | `x >= 7`          | `True` se x for 7 ou mais       |
| `<=`     | Menor ou igual a        | `x <= 4`          | `True` se x for 4 ou menos      |

#### 🔗 Operadores lógicos

Usados para combinar múltiplas condições:

| Operador | Descrição               | Exemplo               | Resultado                                                        |
|----------|------------------------|-----------------------|-----------------------------------------------------------------|
| `and`    | E (todas verdadeiras)  | `(x > 5) and (x < 10)`| `True` se ambas as condições forem verdadeiras                  |
| `or`     | Ou (pelo menos uma verdadeira) | `(x < 3) or (x > 7)`    | `True` se uma ou outra condição for verdadeira                  |
| `not`    | Negação                | `not(x == 5)`         | Inverte o valor lógico (`True` vira `False` e vice-versa)       |

````python
x = 7

# Comparação simples
if x == 7:
    print("x é igual a 7")

# Usando 'and'
if x > 5 and x < 10:
    print("x está entre 6 e 9")

# Usando 'or'
if x < 3 or x > 5:
    print("x é menor que 3 ou maior que 5")

# Usando 'not'
if not x == 5:
    print("x não é 5")
````

### Estruturas de repetição

#### While (enquanto)

O `while` executa um bloco de código **enquanto a condição for verdadeira**.

💡 Use `while` quando você **não sabe exatamente quantas vezes o loop vai rodar**.

````python
import time

i: int = 10

while i >= 0:
    print(i)
    time.sleep(1)  # pausa de 1 segundo
    i -= 1  # mesma coisa que i = i - 1
````

#### For (para)

O `for` é usado quando você **sabe quantas vezes quer repetir algo**.

```python
    for i in range(a, b, c):
        # código
```

* `a` → início (padrão = 0)
* `b` → fim (não incluído)
* `c` → passo (padrão = 1)

````python
# Forma tradicional de usar range:
for i in range(0, 10, 1):
    print(i, end=" ")
    if i == 9:
        print() # quebra de linha após o último número

# Forma simplificada:
for i in range(10):
    print(i, end=" ")
    if i == 9:
        print() # quebra de linha após o último número
````

````python
# Percorrendo lista
nomes = ["Ana", "João", "Maria"]

for nome in nomes:
    print(nome)
````

#### Break e Continue

* `continue` → pula para a próxima iteração
* `break` → encerra o loop

````python
while True:
    if i < 10:
        i += 1
        continue  # volta para o início do loop

    print(i)  # executa quando i == 10
    break  # sai do loop
````

### Estruturas de dados

São formas de **armazenar e organizar dados**.

#### Listas

* Podem armazenar vários valores
* Podem misturar tipos
* Crescem dinamicamente

````python
lista = ["USP", "ICMC", "Python"]

print(lista[0])   # primeiro elemento
print(lista[1])
print(lista[-1])  # último elemento
````

````python
# Adição e remoção de elementos

lista = []

lista.append(1)
lista.append(2)
lista.append("Eugênia")

print(lista)

lista.remove(1)  # remove o valor 1

print(lista)
````

````python
# Outras operações

lista = [3, 2, 1]

print(len(lista))  # tamanho da lista

x = lista.pop(1)  # remove índice 1

print(x)      # 2
print(lista)  # [3, 1]

lista.sort()  # ordena

print(lista)  # [1, 3]
````

💡 Dica: `pop()` retorna o valor removido.

#### Slicing (fatiamento)

Permite pegar partes da lista:

```python
    lista[inicio:fim:passo]
```

````python
lista = list(range(10))  # [0, 1, ..., 9]

print(lista[0:9:1])  # [0, 1, ..., 8]
print(lista[:])      # cópia da lista
print(lista[2:])     # do índice 2 até o final
print(lista[:5])     # do início até índice 4
print(lista[::2])    # de 2 em 2
print(lista[-1])     # último elemento
print(lista[::-1])   # lista invertida
````

#### Set (conjunto)

* Não permite elementos repetidos
* Não mantém ordem
* Uso comum: remover duplicatas

````python
array = [1, 1, 2, 2, 3]

conjunto = set(array)

print(conjunto)  # {1, 2, 3}
````

#### Tupla

* Imutável (não pode ser alterada)
* Mais rápida que lista em alguns casos

Erro comum:

```python
    tupla[0] = 10  # ❌ erro (imutável)
```

````python
tupla = (1, 2, 3)

print(tupla[0])
````

### Dicionário

* Estrutura de **chave → valor**
* Acesso muito rápido: O(1)

````python
dicionario = {}

dicionario[1] = "1"
dicionario["1"] = 1
dicionario["lista"] = [1, 2, 3]
dicionario["dicionario"] = {"key": "valor"}

# Acessando valores
print(dicionario["lista"])  # [1, 2, 3]

# Percorrendo
for chave, valor in dicionario.items():
    print(chave, valor)
````

````python
# Métodos úteis

print(dicionario.keys())   # chaves
print(dicionario.values()) # valores

# verificando existência
if "lista" in dicionario:
    print("Existe!")
````

Exercício 1: Desenvolva um programa que selecione um número aleatório entre 0 e 100. O usuário deverá tentar adivinhar esse número. Ao final, o programa exibirá a sequência de palpites do usuário e o total de tentativas feitas até acertar o número correto

````python
import random

numeroGerado: int = random.randint(0, 100)

numeroUsario: int

Palpites = []

limiteInferior: int = 0
limiteSuperior: int = 100

while True:
    # numeroUsario = int(input("Digite um número entre 0 e 100: ")) # input retorna string, então convertemos para int
    numeroUsario = random.randint(limiteInferior, limiteSuperior)  # simula o palpite do usuário
    Palpites.append(numeroUsario) # adiciona o palpite do usuário à lista de palpites

    if numeroUsario < numeroGerado:
        print("Tente um número maior!")
        limiteInferior = numeroUsario
    elif numeroUsario > numeroGerado:
        print("Tente um número menor!")
        limiteSuperior = numeroUsario
    else:
        print(f"Parabéns! Você acertou em {len(Palpites)} tentativas.")
        break

print("Número gerado:", numeroGerado)
print("Palpites do usuário:", Palpites)
print("Palpites do usuário (usando for):", end=" ")
for i in range(len(Palpites)):
    print(f"{Palpites[i]}", end=" " )
````

#### 📌 Quando usar `f""`

Use f-strings sempre que precisar:

1. Inserir variáveis em uma string

````python
nome = "João"
idade = 25

print(f"Meu nome é {nome} e tenho {idade} anos")
````

2. Avaliar expressões dentro da string

````python
a = 10
b = 5

print(f"A soma é {a + b}")
````

3. Formatar números

````python
preco = 49.9

print(f"Preço: R$ {preco:.2f}")
````

Exercício 2: Você está organizando uma lista de alunos por turmas em uma escola. Cada turma contém vários alunos, e cada aluno é representado por um dicionário com informações sobre seu nome e idade.

a) Criar as turmas: Crie duas turmas, cada uma contendo uma lista de dicionários representando os alunos. Cada dicionário deve ter as chaves "nome" e "idade".

````python
turmas = {}

turmas = {"Matemática I": [{"nome": "Ana", "idade": 9, "id": 1}, {"nome": "João", "idade": 7, "id": 2}], "Matemática II": [{"nome": "Maria", "idade": 10, "id": 3}, {"nome": "José", "idade": 10, "id": 4}]}

print(turmas)
````

b) Adicionar um novo aluno: Adicione um novo aluno em uma das turmas.

````python
novoAluno = {"nome": "Carlos", "idade": 8, "id": 5}

turmas["Matemática I"].append(novoAluno) # adiciona o novo aluno à turma de Matemática I, a turma específica é uma lista
print(turmas)
# imprime mensagem corrigida (com f-string)
print(f"A turma Matemática I agora tem {len(turmas['Matemática I'])} alunos.")
````

c) Remover um aluno: Remova um aluno de uma das turmas.

````python
turmas["Matemática II"].remove(turmas["Matemática II"][0]) # remove o primeiro aluno da turma de Matemática II
print(turmas)

# Outra forma (buscar pelo id do aluno)
id: int = 3

for turma in turmas.keys():
    for nome in turmas[turma]:
        if nome["id"] == id:
            turmas[turma].remove(nome)
            break

print(turmas)
````

Exercício 3: Recrie a pirâmide do Mario

````python
import random

p: int = random.randint(1, 12)
print("Tamanho da base da pirâmide:", p)

# primeiro loop: cria uma "escada" de # alinhada à esquerda
for i in range(p):
    print("#" * (i + 1)) # em python é possível multiplicar strings por inteiros

# segundo loop: cria uma pirâmide alinhada à direita
for i in range(p):
    print(" " * (p - (i + 1)) + "#" * (i + 1))
````
