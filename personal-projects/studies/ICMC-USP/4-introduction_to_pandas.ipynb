### 🔹 1. Introdução ao Pandas

O **Pandas** é uma biblioteca Python que facilita a manipulação de **dados tabulares**.

#### Principais vantagens:

* Conexão com diversas fontes (CSV, Excel, Parquet)
* Estruturas de dados poderosas
* Fácil análise e transformação de dados

### 🔹 2. Tipos de Dados no Pandas

O Pandas usa o **NumPy** como base.

#### Tipos comuns:

* `int`, `float`, `bool`
* `object` → strings, listas, dicionários
* `datetime64` → datas
* `category` → dados categóricos (economiza memória)

### 🔹 3. Series (Coluna)

Uma **Series** é uma estrutura unidimensional (como uma coluna).

#### Componentes:

* Um conjunto de dados unidimensional, que podem ser de qualquer tipo
* Um índice (index) que é usado para acessar cada dado da Series. O índice geralmente é um número inteiro ou uma string (Object)

#### Criando Series:

* Objetos do Python:
    * Listas e Tuplas: Index numérico será gerado de forma automática
    * Dicionários: O index da Series vão ser as Chaves do dicionário
* Arquivos CSV e Excel com apenas uma coluna também podem ser lidos para criar Series.

````python
import pandas as pd

# Lista
serie1 = pd.Series([1, 2, 3])
serie1
````

````python
# Dicionário
serie2 = pd.Series({"A": 1, "B": 2, "C": 3})
serie2
````

### 🔹 14. Merge / Join

Semelhante ao **JOIN do SQL**.

* `on="coluna"`: quando a coluna tem mesmo nome
* `left_on="col_esq"` / `right_on+"col_dir"`: quando os nomes são diferentes
* `how`define o tipo de `join` (`inner`, `left`, `right`, `outer`)

#### Tipos:

* `inner`: traz apenas os registros que existem nos DOIS DataFrames (interseção)
* `left`: traz todos os registros do DataFrame da ESQUERDA + correspondências da direita
* `right`: traz todos os registros do DataFrame da DIREITA + correspondências da esquerda
* `outer`: traz TODOS os registros dos dois DataFrames (união total)

````python
import pandas as pd

# DataFrame principal (funcionários)
df = pd.DataFrame({
    "id": [1, 2, 3, 4],
    "nome": ["Ana", "Bruno", "Carlos", "Diana"],
    "id_departamento": [10, 20, 30, None]
})

# DataFrame de departamentos
df_dep = pd.DataFrame({
    "id_departamento": [10, 20, 40],
    "nome_departamento": ["RH", "TI", "Financeiro"]
})
````

````python
# Só traz quem tem correspondência nos dois
inner_join = pd.merge(df, df_dep, on="id_departamento", how="inner")
print(inner_join)
````

````python
# Traz todos do df e os correspondentes do df_dep (preenchendo com NaN onde não tem correspondência)
left_join = pd.merge(df, df_dep, on="id_departamento", how="left")
print(left_join)
````

````python
# Traz todos do df_dep e os correspondentes do df (preenchendo com NaN onde não tem correspondência)
right_join = pd.merge(df, df_dep, on="id_departamento", how="right")
print(right_join)
````

````python
# Traz todos dos dois data frames, preenchendo com NaN onde não tem correspondência
outer_join = pd.merge(df, df_dep, on="id_departamento", how="outer")
print(outer_join)
````

Exercício 1. Faça uma função que receba um DataFrame e:

* Remova a primeira e a última linha
* Retorne um novo DataFrame com os dados restantes
* Não pode usar drop() diretamente com índice fixo (use lógica com iloc)

````python
def remover_primeira_ultima_linha(df):
    # Verifica se o DataFrame tem pelo menos 3 linhas para evitar erros
    if len(df) < 3:
        raise ValueError("O DataFrame deve ter pelo menos 3 linhas para remover a primeira e a última.")
    
    # Usa iloc para selecionar todas as linhas, exceto a primeira (0) e a última (-1)
    novo_df = df.iloc[1:-1]
    
    return novo_df

# Exemplo de uso
df_exemplo = pd.DataFrame({
    'A': [1, 2, 3, 4, 5],
    'B': [10, 20, 30, 40, 50]
})
print("DataFrame original:")
print(df_exemplo)

df_modificado = remover_primeira_ultima_linha(df_exemplo)
print("DataFrame após remover a primeira e a última linha:")
print(df_modificado)
````

Exercício 3. Crie um DataFrame com números de 0 a 100.

* Filtre apenas os números ímpares
* A partir desse resultado, filtre os números divisíveis por 3 e 5 ao mesmo tempo
* Crie uma nova coluna com a soma acumulada (cumsum)

````python
df_valores = pd.DataFrame({"valores": range(101)})
print(df_valores)
df_pares = df_valores[df_valores['valores'] % 2 == 1]
df_divisiveis = df_pares[(df_pares['valores'] % 3 == 0) & (df_pares['valores'] % 5 == 0)].copy()
df_divisiveis['soma_acumulada'] = df_divisiveis['valores'].cumsum()
print(df_divisiveis)
````

Exercício 4. Faça uma função que receba um DataFrame com coluna "valores" e coluna "pesos"

* Retorne a média ponderada
* NÃO pode usar .mean() com peso nem função pronta
* Use apenas operações matemáticas

````python
def media_ponderada(df):
    # Verifica se as colunas "valores" e "pesos" existem no DataFrame
    if 'valores' not in df.columns or 'pesos' not in df.columns:
        raise ValueError("O DataFrame deve conter as colunas 'valores' e 'pesos'.")
    
    # Calcula o produto dos valores pelos pesos
    produto = df['valores'] * df['pesos']
    
    # Soma dos produtos
    soma_produto = produto.sum()
    
    # Soma dos pesos
    soma_pesos = df['pesos'].sum()
    
    # Calcula a média ponderada
    media_ponderada = soma_produto / soma_pesos
    
    return media_ponderada

# Exemplo de uso
df_exemplo = pd.DataFrame({
    'valores': [10, 20, 30],
    'pesos': [1, 2, 3]
})
resultado = media_ponderada(df_exemplo)
print("Média ponderada:", resultado)
````

Exercício 5. Dado um DataFrame onde cada coluna representa uma prova:

```python
df = pd.DataFrame({
    "prova1": [10, 5, 8],
    "prova2": [7, 9, 6],
    "prova3": [8, 7, 9]
})
```

* Retorne qual prova teve a maior soma de notas

````python
def prova_maior_soma(df):
    # Calcula a soma de cada coluna (prova)
    soma_provas = df.sum()
    
    # Encontra o nome da prova com a maior soma
    prova_maior = soma_provas.idxmax()
    
    return prova_maior

# Exemplo de uso
df_exemplo = pd.DataFrame({
    "prova1": [10, 5, 8],
    "prova2": [7, 9, 6],
    "prova3": [8, 7, 9]
})

prova_com_maior_soma = prova_maior_soma(df_exemplo)
print("A prova com a maior soma de notas é:", prova_com_maior_soma)
````

Exercício 6. Dado um DataFrame:

```python
df = pd.DataFrame({
    "aluno": ["Ana", "Bruno", "Carlos", "Diana"],
    "nota": [7.5, 4.0, 6.0, 3.5]
})
```

* Faça a média das notas
* Maior e menor nota + nome do aluno
* Liste alunos aprovados (nota ≥ 5)
* Liste alunos reprovados

````python
def analise_notas(df):
    # Verifica se as colunas "aluno" e "nota" existem no DataFrame
    if 'aluno' not in df.columns or 'nota' not in df.columns:
        raise ValueError("O DataFrame deve conter as colunas 'aluno' e 'nota'.")
    
    # Calcula a média das notas
    media_notas = df['nota'].mean()
    
    # Encontra a maior nota e o nome do aluno correspondente
    maior_nota = df['nota'].max()
    aluno_maior_nota = df.loc[df['nota'] == maior_nota, 'aluno'].values[0]
    
    # Encontra a menor nota e o nome do aluno correspondente
    menor_nota = df['nota'].min()
    aluno_menor_nota = df.loc[df['nota'] == menor_nota, 'aluno'].values[0]
    
    # Lista de alunos aprovados (nota ≥ 5)
    alunos_aprovados = df[df['nota'] >= 5]['aluno'].tolist()
    
    # Lista de alunos reprovados (nota < 5)
    alunos_reprovados = df[df['nota'] < 5]['aluno'].tolist()
    
    return {
        "media_notas": media_notas,
        "maior_nota": (aluno_maior_nota, maior_nota),
        "menor_nota": (aluno_menor_nota, menor_nota),
        "alunos_aprovados": alunos_aprovados,
        "alunos_reprovados": alunos_reprovados
    }

# Exemplo de uso
df_exemplo = pd.DataFrame({
    "aluno": ["Ana", "Bruno", "Carlos", "Diana"],
    "nota": [7.5, 4.0, 6.0, 3.5]
})
resultados = analise_notas(df_exemplo)
print("Média das notas:", resultados["media_notas"])
print("Maior nota:", resultados["maior_nota"])
print("Menor nota:", resultados["menor_nota"])
print("Alunos aprovados:", resultados["alunos_aprovados"])
print("Alunos reprovados:", resultados["alunos_reprovados"])
````

Exercício 7. Dado:

```python
df = pd.DataFrame({
    "nome": ["Ana", "Bruno", "Carlos", "Diana", "Eduardo"],
    "idade": [23, 35, 29, 41, 22],
    "salario": [2500, 4000, 3200, 5000, 1800]
})
```

* Retorne: apenas as colunas "nome" e "salario" das pessoas com salário > 3000 e idade < 40
* Cre uma coluna "categoria": "Baixo" → salário < 3000; "Médio" → 3000 a 4000; "Alto" → > 4000. Use apply() ou np.where()

````python
def classificar_categoria(salario):
    if salario < 3000:
        return "Baixo"
    elif 3000 <= salario <= 4000:
        return "Médio"
    else:
        return "Alto"

df = pd.DataFrame({
    "nome": ["Ana", "Bruno", "Carlos", "Diana", "Eduardo"],
    "idade": [23, 35, 29, 41, 22],
    "salario": [2500, 4000, 3200, 5000, 1800]
})
df['categoria'] = df['salario'].apply(classificar_categoria)
print(df)
````

Exercício 8. Dado um DataFrame com salários:

* Ordene do maior para o menor salário
* Crie uma coluna "ranking" com a posição de cada funcionário

````python
def ordenar_ranking(df):
    # Ordena o DataFrame por salário (do maior para o menor)
    df_ordenado = df.sort_values("salario", ascending=False).copy()
    
    # Cria a coluna "ranking" com a posição de cada funcionário
    df_ordenado['ranking'] = range(1, len(df_ordenado) + 1)
    
    return df_ordenado

df = pd.DataFrame({
    "nome": ["Ana", "Bruno", "Carlos", "Diana", "Eduardo"],
    "salario": [2500, 4000, 3200, 5000, 1800]
})
df_ranking = ordenar_ranking(df)
print(df_ranking)
````

Exercício 9. Dado:

```python
df = pd.DataFrame({
    "departamento": ["TI", "TI", "RH", "RH", "Financeiro"],
    "salario": [5000, 7000, 4000, 4500, 8000]
})
```

* Faça a média salarial por departamento
* Soma salarial por departamento
* Quantidade de funcionários

````python
df = pd.DataFrame({
    "departamento": ["TI", "TI", "RH", "RH", "Financeiro"],
    "salario": [5000, 7000, 4000, 4500, 8000]
})

media_salarial = df.groupby("departamento")["salario"].mean()
soma_salarial = df.groupby("departamento")["salario"].sum()
quantidade_funcionarios = df.groupby("departamento").size()

print("Média salarial por departamento:")
print(media_salarial)
print("\nSoma salarial por departamento:")
print(soma_salarial)
print("\nQuantidade de funcionários por departamento:")
print(quantidade_funcionarios)
````
