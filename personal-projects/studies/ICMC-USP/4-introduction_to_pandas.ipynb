### 🔹 1. Introdução ao Pandas

O **Pandas** é uma biblioteca Python que facilita a manipulação de **dados tabulares**.

#### Principais vantagens:

* Conexão com diversas fontes (CSV, Excel, Parquet)
* Estruturas de dados poderosas
* Fácil análise e transformação de dados

### 🔹 2. Tipos de Dados no Pandas

O Pandas usa o **NumPy** como base.

#### Tipos comuns:

* `int`, `float`, `bool`
* `object` → strings, listas, dicionários
* `datetime64` → datas
* `category` → dados categóricos (economiza memória)

### 🔹 3. Series

Uma **Series** é uma estrutura unidimensional (como uma coluna).

#### Componentes:

* Um conjunto de dados unidimensional, que podem ser de qualquer tipo
* Um índice (index) que é usado para acessar cada dado da Series. O índice geralmente é um número inteiro ou uma string (Object)

#### Criando Series:

* Objetos do Python:
    * Listas e Tuplas: Index numérico será gerado de forma automática
    * Dicionários: O index da Series vão ser as Chaves do dicionário
* Arquivos CSV e Excel com apenas uma coluna também podem ser lidos para criar Series.

```python
import pandas as pd

# Lista
serie1 = pd.Series([1, 2, 3])

# Dicionário
serie2 = pd.Series({"A": 1, "B": 2, "C": 3})

print(serie1)
print(serie2)
```

### 🔹 4. DataFrames

Um **DataFrame** é uma tabela (linhas e colunas).

* Cada coluna é uma Series

### 🔹 9. View vs Copy

```python
# View (mesmo objeto)
view_df = df[["col1", "col2"]]
view_df.at[0, "col1"] = 999  # altera o original

# Copy (novo objeto)
copy_df = df[["col1", "col2"]].copy()
copy_df.at[0, "col1"] = 999  # NÃO altera o original
```

### 🔹 10. Métodos de Series

#### Estatísticos

```python
# Estatísticos
df["col1"].mean()     # média dos valores
df["col1"].median()   # mediana (valor central)
df["col1"].mode()     # valor(es) mais frequente(s)
df["col1"].min()      # menor valor
df["col1"].max()      # maior valor

# Aritméticos
df["col1"].add(10)    # soma 10 a cada elemento
df["col1"].sub(5)     # subtrai 5 de cada elemento
df["col1"].mul(2)     # multiplica cada elemento por 2
df["col1"].div(2)     # divide cada elemento por 2

# Gerais
df["col1"].count()                 # conta valores não nulos
df["col1"].value_counts()          # conta frequência de cada valor
df["col1"].apply(lambda x: x * 2)  # aplica função em cada elemento
```

### 🔹 11. Merge / Join

Semelhante ao **JOIN do SQL**.

* `on="coluna"`: quando a coluna tem mesmo nome
* `left_on="col_esq"` / `right_on+"col_dir"`: quando os nomes são diferentes
* `how`define o tipo de `join` (`inner`, `left`, `right`, `outer`)

#### Tipos:

* `inner`: traz apenas os registros que existem nos DOIS DataFrames (interseção)
* `left`: traz todos os registros do DataFrame da ESQUERDA + correspondências da direita
* `right`: traz todos os registros do DataFrame da DIREITA + correspondências da esquerda
* `outer`: traz TODOS os registros dos dois DataFrames (união total)

### 🔹 12. GroupBy

Agrupa dados por valores únicos.

* `groupby()`: define os grupos
* Selecionar coluna: define o que será analisado
* Agregação: resume os dados de cada grupo

```python
# Agrupa o DataFrame pelos valores únicos da coluna "cidade"
agrupamento = df.groupby("cidade")
```

```python
# Seleciona a coluna "faturamento" dentro do agrupamento
# Agora temos um objeto agrupado só dessa coluna
faturamento = agrupamento["faturamento"]
```

```python
# Aplica funções de agregação para cada grupo (cada cidade)

faturamento.sum()    # soma total por cidade
faturamento.mean()   # média por cidade
faturamento.max()    # maior valor por cidade
faturamento.count()  # quantidade de registros por cidade
```
